Getting the Dataset

```python
import pandas as pd
import seaborn as sns

books = pd.read_csv("data/BX-Books.csv", sep=";", encoding="latin-1")
users = pd.read_csv("data/BX-Users.csv", sep=";", encoding="latin-1")
ratings = pd.read_csv("data/BX-Book-Ratings.csv", sep=";", encoding="latin-1")
```
